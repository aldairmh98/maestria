```python
"""

Parte 2

Con lo aprendido en clase, elige uno de los siguientes problemas e implementa un programa en Python utilizando programación orientada a objetos. La solución debe incluir una clase personalizada con al menos dos métodos.

Opción B. Diccionario musical

Crea una clase llamada Cancion que permita organizar la información de una canción.

La clase debe incluir los siguientes atributos:

    artista
    álbum
    duración
    género

Además, debe incluir al menos los siguientes métodos:

    crear_diccionario(): genera un diccionario con la información de la canción.
    mostrar_info(): muestra las claves (keys), los valores (values) y las tuplas (items) del diccionario generado.

"""
```

```python
class Cancion:
    artista: str
    album: str
    duracion: str
    genero: str
    nombre: str

    def __init__(self,
                artista: str,
                album: str,
                duracion: str,
                genero: str,
                nombre: str
                 ):
        self.artista = artista
        self.album = album
        self.duracion = duracion
        self.genero = genero
        self.nombre = nombre

    def crear_diccionario(self):
        return {
            'artista': self.artista,
            'album': self.album,
            'duracion': self.duracion,
            'genero': self.genero
        }

    def mostrar_info(self):
        diccionario = self.crear_diccionario()
        _keys = diccionario.keys()
        print(_keys)
        _values = diccionario.values()
        print(_values)
        _tuplas = diccionario.items()
        print(_tuplas)
```

```python
cancion = Cancion('Michael Jackson', 'Thriller', '4:18', 'Pop', 'Beat it')
cancion.mostrar_info()
```
